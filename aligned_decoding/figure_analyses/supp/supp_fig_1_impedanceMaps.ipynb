```python
from mpl_toolkits.axes_grid1 import make_axes_locatable
def chan_disp(data, chan_map, title, label, figsize=(8, 6), cbar_lower=None, cbar_upper=None, cmap=None, return_fig=False, show_plot=True):

    mean_val = np.nanmean(data)
    std_val = np.nanstd(data)
    std_mult = 3
    # clim = [max(0, mean_val - std_mult*std_val), mean_val + std_mult*std_val]
    if cbar_lower is None:
        cbar_lower = mean_val - std_mult*std_val
    if cbar_upper is None:
        cbar_upper = mean_val + std_mult*std_val
    clim = [cbar_lower, cbar_upper]

    if cmap is None:
        cmap = parula_map

    chan_data = np.array([data[int(i)] if not np.isnan(i) else np.nan for i in (chan_map.flatten()-1)])
    chan_data = chan_data.reshape(chan_map.shape)
    f = plt.figure(figsize=figsize)
    im = plt.imshow(chan_data, cmap=parula_map)
    im.set_clim(clim[0], clim[1])
    # plt.gca().invert_yaxis()
    # plt.contourf(chan_data, cmap='viridis')
    plt.xticks([])
    plt.yticks([])
    plt.title(title)

    divider = make_axes_locatable(plt.gca())
    cax = divider.append_axes('right', size='5%', pad=0.05)

    cbar = plt.colorbar(im, cax=cax)
    cbar.set_label(label)
    cbar.outline.set_visible(False)

    sns.despine(left=True, bottom=True)

    if show_plot:
        plt.show()
    
    if return_fig:
        return f
```

# Load and plot impedance maps

```python
pt_names = ['S14', 'S26', 'S33', 'S22', 'S23', 'S39', 'S58', 'S62']
# pt_names = ['S14', 'S26', 'S22', 'S23', 'S39', 'S58', 'S62']
data_dir = Path('~\Box\CoganLab\Data\Micro\BIDS_processing').expanduser()
raw_fname = 'sub-%s/sub-%s_raw.h5'
imped_fname = 'sub-%s/sub-%s_impedance.npy'
```

```python
fig_dir = Path(r'~\Box\CoganLab\Papers\2025\cross-patient-speech-decoding\figures\supp\impedance_maps').expanduser()
```

```python
for pt in pt_names:
    impedance_data = np.load(data_dir / (imped_fname % (pt,pt)))
    good_chans = np.where(impedance_data < 1e6)[0]
    with h5py.File(data_dir / (raw_fname % (pt,pt)), 'r') as f:
        channel_map = f['channel_map'][:]

    if pt in ['S14', 'S22', 'S23', 'S26']:
        channel_map = np.fliplr(channel_map.T)
    else:
        channel_map = np.fliplr(np.flipud(channel_map))

    fig = chan_disp(impedance_data/1e6, channel_map, title=f'{paper_pt_dict[pt]} - Mean Impedance: {np.mean(impedance_data[good_chans]/1e3):.1f} kOhm', label='Impedance (MOhm)', figsize=(8, 6), cbar_lower=0, cbar_upper=1, cmap=parula_map, return_fig=True)
    fig.savefig(fig_dir / f'{pt}_impedance_map.svg', dpi=300, bbox_inches='tight', format='svg')
```
